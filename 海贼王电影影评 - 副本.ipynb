```python
import pandas as pd
import numpy as np
import requests
import bs4
movie_encoding={'红发歌姬 ':'35675082',
     '强者天下':'3915458',
     '剧场版Z':'10467125',
     '狂热行动':'30329892',
     '航海王之黄金城':'26598021',
     '强者世界前传':'4837828',
     '另一名同伴的故事':'25711544',
     '冬季绽放的奇迹之樱 ':'3017508',
     '三兄弟的羁绊':'26438933',
     '黄金岛冒险':'1856675',
     '空岛之章':'30313854',
     '死亡尽头的冒险 ':'2279997',
     '狂欢男爵和神秘岛':'3263410',
     '海贼王&龙珠超 合作秋季特番':'27126343',
     '3D2Y 跨越艾斯之死':'25975887',
     '黄金之心':'26834930',
     '沙漠公主与海盗们':'2085498',
     '珍兽岛的乔巴王国':'2252594',
     '航海士的眼泪与牵绊':'11636325',
     '迷雾岛大冒险':'26692504',
     '发条岛的冒险':'1856674',
     '绝妙孤岛':'24845121',
     '被诅咒的圣剑':'2280004',
     '草帽小子海贼团的秘密':'5339105',
     '机关城的机械巨兵':'1860453'
     }
```

```python
### 一个电影的
def oneview(url,n):
    kv={'user-agent':'Mozilla/5.0'}  #模拟浏览器来爬取！
    divlist=[]
    try:
        for i in range(int(n)):
            a=i*20
            url1=url+'?start='+str(a)
            r=requests.get(url1,headers=kv)
            soup=bs4.BeautifulSoup(r.text,'html.parser')
            div=soup.find_all('div',class_='short-content')
            divlist=divlist+[[j for j in i.get_text('...',strip=True).split('...') if j!='这篇影评可能有剧透'][0] for i in div]#separator='@',分割出我们要的内容
        return divlist
    except requests.exceptions.RequestException as e:
        return f"请求发生异常: {e}"

    except:
        return '超出评论页数'
#获取不同电影的url
def page(m):
    try:
        for i in list(movie_encoding.values())[:int(m)]:
            url='https://movie.douban.com/subject/'+str(i)+'/reviews'
            yield url
    except requests.exceptions.RequestException as e:
        return f"请求发生异常: {e}"
    except:
        return '超出电影数量'
#把一个电影的评论变为series
def filldataframe(divlist,m,count):
    try:
        df1=pd.Series(data=divlist).rename(list(movie_encoding.keys())[count])
        return df1
    except requests.exceptions.RequestException as e:
        return f"请求发生异常: {e}"
def main():
    df = []
    n = input('爬取每个电影评论页数：')
    m = input('爬取电影数量：')
    count=0
    
    try:
        for url in page(m):
            df2 = filldataframe(oneview(url, n), m,count)
            df.append(df2)
            dataframe = pd.concat(df, axis=1)
            count=count+1
            print('\r当前总进度：{:.2f}%'.format(count*100/int(m)),end='')#打印进度
        return dataframe
    except:
        return ''
main()
```
